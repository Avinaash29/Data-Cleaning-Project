```python
#using forwardfill to fill the missing values in the order date
df["Order_Date"].fillna(method="ffill",inplace=True)
df
```

```python
#using forwardfill to fill the missing values in the customer name
df["Customer_Name"].fillna(method="ffill",inplace=True)
df
```

```python
#using backwardfill to fill the missing values in the product category
df["Product_Category"].fillna(method="bfill",inplace=True)
df
```

```python
#using fillna to fill the missing value as 2 in quantity
df["Quantity"].fillna(2,inplace=True)
df
```

```python
#using mean to fill the missing values in unit price
df["Unit_Price"].fillna(df['Unit_Price'].mean(),inplace=True)
df
```

```python
# multiple the quantity and unit price to find the total price
df["Total_Price"]=df["Quantity"] * df["Unit_Price"]
df
```

```python
#using backwardfill to fill the missing values in the shipping status
df["Shipping_Status"].fillna(method="bfill",inplace=True)
df
```

```python
#using dropna for remove missing values
df.dropna(inplace=True)
df
```

```python
#to check weather there is a duplicates
df.duplicated().sum()
```

```python
# to remove duplicates
df.drop_duplicates(inplace=True)
df
```

```python
#rename the column name
df.rename(columns={"City":"location"},inplace=True)
df
```

```python
#resetting the index
df.reset_index(drop=True,inplace=True)
df
```

```python
# set column name as index
df.set_index("Order_ID")
```

```python
#sorting the product category (A to Z)
df.sort_values(by="Product_Category")
```
